```python
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel

!pip install pyLDAvis

import pyLDAvis
import pyLDAvis.gensim
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sent_to_words(sentences, deacc = True):
  for sentence in sentences:
    yield(gensim.utils.simple_preprocess(str(sentence))) # tokenize and preprocess each sentence.

    # individual words, and converts them to lowercase. 
    # It also removes punctuation marks, and filters out any words 
    # that are less than two characters long or are considered stop words 

d = reviews.values.tolist()
d_words = list(sent_to_words(d))

# preprocessed tokens for each sentence in the original reviews list.
print(d_words[3])
```

```python
id2word = corpora.Dictionary(d_words) #dictionary where each unique word is assigned a unique numerical ID
corpus = [id2word.doc2bow(text) for text in d_words] #preprocessed sentence into a 
# bag-of-words representation, where each word is replaced with its corresponding ID
# corpus = [ [(4, 1), (2, 1), (0, 1), (3, 1)],
#  [(4, 1), (2, 1), (1, 1), (3, 1)]
# ]
```

# Building the topic model

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus = corpus,
                                            id2word = id2word,
                                            num_topics = 10,
                                            random_state = 100,
                                            update_every=1,
                                            chunksize = 100,
                                            passes = 10,
                                            alpha = 100,
                                            per_word_topics = True)
```

```python
pprint.pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
# coherence score
coherence_model_lda = CoherenceModel(model = lda_model,
                                     texts = d_words,
                                     dictionary = id2word,
                                     coherence = "c_v")
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score:',coherence_lda)
```

```python
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model,corpus,id2word)
vis
```

# Using bigrams to get topics

```python
def bigrams(words, bi_min = 15, tri_min = 10):
  bigram = gensim.models.Phrases(words, min_count = bi_min)
  bigram_mod = gensim.models.phrases.Phraser(bigram)
  return bigram_mod
```

```python
bigram_mod = bigrams(d_words)
```

```python
bigram = [bigram_mod[review] for review in d_words]
id2word = gensim.corpora.Dictionary(bigram)
id2word.filter_extremes(no_below=10,no_above = 0.35)
id2word.compactify()
corpus = [id2word.doc2bow(text) for text in bigram]
```

```python
# build LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus = corpus,
                                            id2word = id2word,
                                            num_topics = 10,
                                            random_state = 100,
                                            update_every=1,
                                            chunksize = 100,
                                            passes = 10,
                                            alpha = 100,
                                            per_word_topics = True)
```

```python
pprint.pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
#  coherence score
coherence_model_lda = CoherenceModel(model = lda_model,
                                     texts = d_words,
                                     dictionary = id2word,
                                     coherence = "c_v")
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score:',coherence_lda)
```

```python
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model,corpus,id2word)
vis
```
