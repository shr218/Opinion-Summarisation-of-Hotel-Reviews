```python
import pandas as pd
import ast
import re
import nltk
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
df= pd.read_csv(path)
```

```python
df.head()
```

```python
room_positive_values = df.loc[df['aspect'] == 'room', 'positive']
```

```python
for index, value in room_positive_values.iteritems():
    temp=value
reviews= ast.literal_eval(temp)
```

```python
corpus=""
for i in reviews:
  print(i)
  corpus+= i+' '
```

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("t5-base")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-base")
phrase_extraction = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

phrases = []

for review in reviews:
    phrase = phrase_extraction(review, max_length=30, num_return_sequences=1, do_sample=False)
    phrases.append(phrase[0]['generated_text'].strip())

print(phrases)
```

```python
phrase2=[]
for i in phrases:
  temp = re.split(r'[.|,]', i)
  phrase2.extend(temp)
print(phrase2)
```

```python
final=[]
for i in phrase2:
  if i=='':
    continue
  words = nltk.word_tokenize(i)
  tagged_words = nltk.pos_tag(words)
  for word, tag in tagged_words:
      if tag == 'PRP$' or tag == 'PRP':
        continue
  if ('my' in i):
    continue
  if i.strip().lower() not in final:
    final.append(i.strip().lower())
```

```python
print("\n\nOpinion phrases:\n")
for i in final:
  print(i)
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

model_name = 'gpt2'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Load the corpus of hotel reviews
#corpus_file = 'corpus.txt'
#with open(corpus_file, 'r', encoding='utf-8') as f:
#    corpus = f.read()

encoded_corpus = tokenizer.encode(corpus)

model.train()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-5)
for i in range(100):
    loss = model(torch.tensor(encoded_corpus).unsqueeze(0), labels=torch.tensor(encoded_corpus).unsqueeze(0)).loss
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

# Generate text based on the given phrases
opinions = ['Fantastic beautiful', 'own bedouin', 'huge beds', 'exotic bathrooms']
for opinion in opinions:
    input_ids = tokenizer.encode(opinion, return_tensors='pt')
    generated_text = model.generate(input_ids=input_ids, max_length=200, num_return_sequences=1)

    generated_text = [tokenizer.decode(g, skip_special_tokens=True) for g in generated_text]

    print(f"{opinion}: {generated_text[0]}")
```
